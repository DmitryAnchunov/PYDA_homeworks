```python
import glob
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim import corpora
from gensim.models import TfidfModel
from gensim import similarities
import seaborn as sns
from scipy.cluster import hierarchy
```

```python
spam_files = pd.read_csv('spam.csv')
spam_files_copy = spam_files.copy()
spam_files_copy.head()
```

```python
# приведем все слова к нижнему регистру
spam_files['Message'] = spam_files.Message.str.lower()
spam_files.head()
```

```python
# удалим мусорные символы
spam_files['Message'] = spam_files.Message.replace('[\W_]+',' ', regex=True)
spam_files.head()
```

```python
# удалим стоп-слова
stop = set(stopwords.words('english'))
spam_files['Message'] = spam_files.Message.apply(lambda x: [word for word in x.split() if word not in stop])
spam_files.head()
```

```python
# приведем все слова к нормальной форме
wordnet_lemmatizer = WordNetLemmatizer()
spam_files['Message'] = spam_files.Message.apply(lambda x: ' '.join([wordnet_lemmatizer.lemmatize(word) for word in x]))
```

```python
spam_files.head()
```

```python
# преобразуем все сообщения в вектора TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam_files.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
tfidf_matrix['Category'] = spam_files.Category
tfidf_matrix.head()
```

```python
# разделим данные на тестовые и тренировочные в соотношении 30/70, укажите random_state=42. 
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(tfidf_matrix, test_size=0.3, random_state=42)
```

```python
data_train
```

```python
# удалим из тестовой и тренировочной выборок категорию
y_train, y_test = data_train['Category'], data_test['Category']
del data_train['Category']
del data_test['Category']
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
# построим модель логистической регрессии, при random_state=42
clf = LR.fit(data_train, y_train)
clf
```

```python
# оценим точность модели на тестовых данных
print(clf.score(data_test, y_test))

divination = clf.predict(data_test)
result = pd.DataFrame([y_test.values, divination]).T
result.head()
```

```python
# опишем результаты при помощи confusion_matrix
from sklearn.metrics import confusion_matrix

tn, fp, fn, tp = confusion_matrix(y_test.values, divination).ravel()
tn, fp, fn, tp
```
