### Задание 2. Работа с выбросами

1. Найти выбросы 

2. Выдвинуть гипотезы об их причинах и проинтерпретировать результаты

3. Принять и обосновать решение о дальнейшей работе с ними

```python
def quantiles(column):
    
    maxim = new_df[column].max()
    minim = new_df[column].min()
    q1 = new_df[column].quantile(0.25)
    q3 = new_df[column].quantile(0.75)
    iqr = q3 - q1
    if iqr == 0:
        iqr = 0.1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    
    return q1, lower_bound, minim, q3, upper_bound, maxim, iqr
```

```python
def presence_outliers(column):
    
    if (quantiles(column)[5] > quantiles(column)[4]) or (quantiles(column)[2] < quantiles(column)[1]):
        
        return True
    else:
        return False
    
    
```

```python
def list_outliers(column):
    
    list_outliers_low = new_df[new_df[column].between(0, quantiles(column)[1], inclusive=True)]
    list_outliers_up = new_df[new_df[column].between(quantiles(column)[4], quantiles(column)[5], inclusive=True)]

    list_outliers = pd.concat([list_outliers_low, list_outliers_up])
    
    return list_outliers.sort_values(column)
    
```

```python
colums_outliers = []

for column in colums:
    if presence_outliers(column):
        colums_outliers.append(column)
colums_outliers
```
