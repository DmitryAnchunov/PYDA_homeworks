#### Уровень 0

Посчитайте значение 1-ой и 2-ой производной фунции:
f(x) = x^5 + 4*sin(2x) + cos(3x+3) в точке x = 1

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx=1e-6) # значение 1 производной в точке х = 1
```

```python
x, y = symbols('x y')
diff(x**5 + 4*sin(2*x) + cos(3*x + 3))
```

```python
def f1(x): # функция 1 производной
    return 5*(x**4) - 3*sin(3*x + 3) + 8*cos(2*x)
derivative(f1, 1.0, dx=1e-6) # значение 2 производной в точке х = 1
```

```python
a = np.linspace(-5, 5)
fx = [f(xx) for xx in a]
f1x = [f1(y) for y in a]

plt.plot(a, fx, label='function')
plt.plot(a, f1x, label='derivative')
plt.legend()
plt.show()
```
