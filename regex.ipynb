## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
def check_validity_tr_numb(car_id):
    import re
    if re.findall(r'([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})', car_id):
        check = re.findall(r'([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})', car_id)
        print('Номер', check[0][0], 'валиден.', 'Регион:', check[0][1])
    else:
        print('Номер не валиден')
````

## Задание 2

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат: 
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
def deleting_repeated_words(some_string):
    import re
    ss = some_string
    for i in re.findall(r'\w+', some_string):
        count = ss.count(i)
        if count > 1:
            pattern = re.compile(r'(\s{0})+'.format(i))
            i = ' ' + i
            ss = re.sub(pattern, i, ss)
    print(ss)
````

````python
some_string = 'Напишите функцию функцию, которая будет будет будет \
будет удалять все все все все последовательные повторы слов из из из из \
заданной строки строки при помощи регулярных выражений'
````

````python
deleting_repeated_words(some_string)
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
def acronym(some_words):
    import re
    acronym = ''.join(re.findall(r'\b\w', some_words))
    return acronym.upper()
````

````python
some_words = 'Near Field Communication'
acronym(some_words)
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
def numberof_domain_zones(emails):
    import re
    list_d = re.findall(r'(?<=@)\w+\.\w+', str(emails))
    dict_ = {}
    for i in list_d:
        dict_[i] = list_d.count(i)
    [print("{0}: {1}".format(key,value)) for key, value in dict_.items()]
````

````python
emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru, xyz@gmail.com']
````

````python
numberof_domain_zones(emails)
````

Пример работы программы:  

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'`

Результат: 
```
Слов на гласные буквы: 9
Слов на согласные буквы: 21 
```

````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, \
поддерживающий большинство активно используемых платформ. \
Он распространяется под свободной лицензией Python Software Foundation \
License, позволяющей использовать его без ограничений в любых приложениях, \
включая проприетарные.'
````

````python
def count_words_type_sounds(some_text):
    import re
    list_vowel = re.findall(r'\b[иеоуаыэёюяaeiouy]', some_text.lower())
    list_consonant = re.findall(r'\b[^иеоуаыэёюяaeiouy]', some_text.lower())
    print('Слов на гласные буквы:', len(list_vowel))
    print('Слов на согласные буквы:', len(list_consonant))
````

````python
count_words_type_sounds(some_text)
````
