### Центральная предельная теорема¶

В этом задании нужно убедиться, что ЦПТ действительно работает.

Создайте случайную величину из любого выбранного вами распределения, 

для разных значений n сгенерируйте 1000 выборок размера n и 

постройте гистрограммы средних этих выборок.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 1.0
sigma = 0.5

# зададим нормально распределенную случайную величину
rv = sts.norm(loc=mu, scale=sigma)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)

plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
# YOUR CODE HERE

samples_count = 1000
#массивы для сохранения средних по выборкам
a1 = np.array([])

for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    n1 = n1 + number
    sample1 = rv.rvs(size=n1)
    
    # считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    if n1 != 1000:
        n1 = 2
    else:
        break
```

```python
#строим графики
#подписи для графиков
label_1 = 'sample of ' + str(n1) + ' histogram'
plt.hist(a1, bins=500, density=True, label=label_1)

plt.legend(frameon=False, loc='upper center', ncol=1)
plt.ylabel('samples means')
plt.xlabel('$x$')
```
