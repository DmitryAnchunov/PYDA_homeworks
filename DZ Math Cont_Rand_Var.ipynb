#### –ó–∞–¥–∞–Ω–∏–µ 1

–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ç—É–∏—Ü–∏—é, –Ω–∞ —á—Ç–æ –≤–ª–∏—è—é—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π. 

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ 5 —Ñ—É–Ω–∫—Ü–∏–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: 

–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ, –≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ, –°—Ç—å—é–¥–µ–Ω—Ç–∞ –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

### –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ scipy.stats:

loc =  $\mu$, –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ

scale = ùúé, –¥–∏—Å–ø–µ—Ä—Å–∏—è, —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ


```python
grid = np.linspace(-7, 7, 500) # —Å–µ—Ç–∫–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
loc_values = [0, 0, 2, 2, 4] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúá
scale_values = [4, 2, 2, 0.5, 0.5] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúé

plt.figure(figsize=(12, 6))

for i, (lc, sc) in enumerate(zip(loc_values, scale_values)):
 plt.plot(grid, sts.norm(lc, sc).pdf(grid), lw=3,
 label='$\mathcal{Param}' + '({}, {})$'.format(lc, sc))

plt.legend(fontsize=16)
plt.title('–ü–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=20)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ', fontsize=18)
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=18)
plt.show()
```

### –≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ scipy.stats:

loc =  —Å–¥–≤–∏–≥

scale = –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ


```python
grid = np.linspace(0,5, 100, 500) # —Å–µ—Ç–∫–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
loc_values = [0.5, 0.5, 1, 1, 2] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ loc
scale_values = [5, 3, 3, 2, 2] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ scale

plt.figure(figsize=(12, 6))

for i, (lc, sc) in enumerate(zip(loc_values, scale_values)):
 plt.plot(grid, sts.expon(lc, sc).pdf(grid), lw=3,
 label='$\mathcal{Param}' + '({}, {})$'.format(lc, sc))

plt.legend(fontsize=16)
plt.title('–ü–ª–æ—Ç–Ω–æ—Å—Ç–∏ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=20)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ', fontsize=18)
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=18)
plt.show()
```

### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –°—Ç—å—é–¥–µ–Ω—Ç–∞

–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ scipy.stats:

df = –ø–∞—Ä–∞–º–µ—Ç—Ä —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã

loc =  —Å–¥–≤–∏–≥

scale = –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ


```python
grid = np.linspace(0,5, 100, 500) # —Å–µ—Ç–∫–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
df_values = [2, 5, 5, 5, 50] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ df
loc_values = [3, 3, 1, 1, 1] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ loc
scale_values = [0.5, 0.5, 0.5, 1, 1] # –Ω–∞–±–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ scale

plt.figure(figsize=(12, 6))

for i, (df, lc, sc) in enumerate(zip(df_values, loc_values, scale_values)):
 plt.plot(grid, sts.t(df, lc, sc).pdf(grid), lw=3,
 label='$\mathcal{Param}' + '({}, {}, {})$'.format(df, lc, sc))

plt.legend(fontsize=16)
plt.title('–ü–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞', fontsize=20)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ', fontsize=18)
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=18)
plt.show()
```
